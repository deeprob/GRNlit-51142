# Model Elements 

The model has the following elements:

1. Transcription Factors - denoted by their diamond shape
2. Target Genes - denoted by their elliptical shape
3. Transcription factor cluster - denoted by their hexagonal shape
4. Target Gene cluster - denoted by their octagonal shape
5. A metabolic process - denoted by their rectangular shape
